```python
import pandas as pd
import numpy as np
data=pd.read_csv('titanic.csv')
data
```

```python
data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1,inplace=True)
```

```python
data
```

```python
inputs=data.drop(['Survived'],axis=1)
target=data.Survived
```

```python
inputs
```

```python
target
```

```python
#Data_preprocessing
```

```python
dummies=pd.get_dummies(data.Sex)
inputs=pd.concat([inputs,dummies],axis=1)
```

```python
inputs
```

```python
inputs=inputs.drop(["Sex"],axis=1)
```

```python
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.columns[inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(inputs,target,test_size=.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
model.predict_proba(xtest)
```
