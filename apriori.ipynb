```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data=pd.read_csv("OnlineRetail.csv_.zip",encoding='unicode_escape')
```

```python
data
```

```python
data['InvoiceNo']=data['InvoiceNo'].astype('str')
```

```python
data
```

```python
data=data[~data['InvoiceNo'].str.contains('c')]
```

```python
data
```

```python
basket_France=(data[data['Country']=="France"].groupby(["InvoiceNo","Description"]))
```

```python
basket_France
```

```python
for i,j in basket_France:
    print(i)
    print(j)
```

```python
basket_France=(data[data['Country']=="France"].groupby(["InvoiceNo","Description"])).sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
```

```python
basket_France
```

```python
basket_Por=(data[data['Country']=="Portugal"].groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
```

```python
basket_Sweden=(data[data['Country']=="Sweden"].groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
```

```python
basket_UK=(data[data['Country']=="UK"].groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
```

```python
def hot_encode(x):
    if(x<=0):
        return 0
    if(x>=1):
        return 1
```

```python
basket_encoded=basket_France.applymap(hot_encode)
basket_France=basket_encoded
```

```python
basket_encoded
```

```python
basket_encoded=basket_UK.applymap(hot_encode)
basket_UK=basket_encoded
basket_encoded=basket_Por.applymap(hot_encode)
basket_Por=basket_encoded
basket_encoded=basket_Sweden.applymap(hot_encode)
basket_Sweden=basket_encoded
```

```python
freq=apriori(basket_France,min_support=0.05,use_colnames=True)
freq
```

```python
rules=association_rules(freq,metric='lift',min_threshold=1)
```
