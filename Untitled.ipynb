```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
train_data["Cabin"]=train_data["Cabin"].fillna("Zero",inplace=True)
test_data["Cabin"]=test_data["Cabin"].fillna("Zero",inplace=True)
```

```python
m=train_data.Age.mean()
train_data.Age=train_data.Age.fillna(m)
test_data.Age=test_data.Age.fillna(m)
```

```python
test_data["Fare"] = test_data["Fare"].fillna(test_data['Fare'].mean())
train_data = train_data.dropna(subset=["Embarked"])
```

```python
dummies=pd.get_dummies(train_data.Sex)
dummies1=pd.get_dummies(test_data.Sex)
```

```python
train_data=pd.concat([train_data,dummies],axis=1)
train_data=train_data.drop(['Sex'],axis=1)
test_data=pd.concat([test_data,dummies1],axis=1)
test_data=test_data.drop(['Sex'],axis=1)
```
